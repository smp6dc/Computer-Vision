```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

#Implementation of OTSU Approach ::  Task 1 ::

def OTSU(img):
    imageX = cv2.GaussianBlur(img, (5, 5), 0)
    bins_num = 256
    hist, bin_edges = np.histogram(imageX, bins=bins_num)
    hist = np.divide(hist.ravel(), hist.max())
    bin_mids = (bin_edges[:-1] + bin_edges[1:]) / 2.
    weight1 = np.cumsum(hist)
    weight2 = np.cumsum(hist[::-1])[::-1]
    mean1 = np.cumsum(hist * bin_mids) / weight1
    mean2 = (np.cumsum((hist * bin_mids)[::-1]) / weight2[::-1])[::-1]
    inter_class_variance = weight1[:-1] * weight2[1:] * (mean1[:-1] - mean2[1:]) ** 2
    index_of_max_val = np.argmax(inter_class_variance)
    threshold = bin_mids[:-1][index_of_max_val]
    return threshold


def BinImgOP(threshold, pixelEditor, image1):
    w,h = image1.size
    for a in range(0,w):
        for b in range(0,h):
            grayValue = pixelEditor[a,b]
            if grayValue > threshold:
                pixelEditor[a,b] = 256
            else:
                pixelEditor[a, b] = 0                
                
in_A = int(input("Enter 1- coins.bmp image\nEnter 2- lena.png image\nEnter 3- image1.jpg\nEnter 4- image2.jpg\n"))


if(in_A == 1):
    image_path = 'coins.bmp'

elif(in_A == 2):
    image_path = 'lena.png'
    
elif(in_A == 3):
    image_path = 'image1.jpg'
    
elif(in_A == 4):
    image_path = 'image2.jpg'
    
else:
    print("Please Enter Valid Input Value")
    exit(1)

image1 = Image.open(image_path).convert('L')
pixel_Edit = image1.load()
w,h = image1.size
image = np.asarray(image1)

threshold = OTSU(image)

print("Otsu's algorithm implementation thresholding result: ", round(threshold))
threshold =int(threshold)
BinImgOP(threshold, pixel_Edit, image1)
image1.show()

#Implementation of APT Approach :: Task 2 ::

imgAPT = Image.open(image_path).convert('L')
imgAPT = imgAPT.quantize(threshold)
pixelEditor = imgAPT.load()

imgAPT = np.asarray(imgAPT)

threshold = OTSU(imgAPT)
threshold = int(threshold)
imgAPT = Image.open(image_path).convert('L')
pixelEditor = imgAPT.load()
BinImgOP(threshold, pixelEditor, imgAPT)

print('APT algorithm implementation thresholding result:', threshold)
BinImgOP(threshold, pixelEditor, imgAPT)
imgAPT.show()
```
